```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import FancyBboxPatch

def create_clean_schematic():
    fig, ax = plt.subplots(figsize=(12, 8))

    # Define key levels and their respective steps
    steps = {
        "Data Preparation": [
            "Load Libraries and Data",
            "Aggregate Project Data by Address"
        ],
        "Map Setup": [
            "Initialize Map Container",
            "Highlight Countries with Projects"
        ],
        "Symbol and Styling": [
            "Assign Dynamic Symbols",
            "Configure Geo and Scatter Plot"
        ],
        "Interactivity": [
            "Add Hover Tooltips",
            "Enable Click Events"
        ],
        "Final Rendering": [
            "Render Map"
        ]
    }

    # Set vertical positions for each level
    levels = list(steps.keys())
    level_positions = [0.9 - i * 0.2 for i in range(len(levels))]

    # Draw the schematic flowchart
    for i, (level, pos_y) in enumerate(zip(levels, level_positions)):
        # Draw level box
        ax.add_patch(FancyBboxPatch(
            (0.05, pos_y - 0.04), 0.9, 0.08,
            boxstyle="round,pad=0.2", edgecolor="black", facecolor="lightgray"
        ))
        ax.text(0.5, pos_y, level, fontsize=12, fontweight="bold", ha="center", va="center")

        # Add steps
        step_y = pos_y - 0.1
        for j, step in enumerate(steps[level]):
            x = 0.15 + j * 0.3  # Space out steps horizontally
            ax.add_patch(FancyBboxPatch(
                (x - 0.15, step_y - 0.04), 0.3, 0.08,
                boxstyle="round,pad=0.2", edgecolor="black", facecolor="lightblue"
            ))
            ax.text(x, step_y, step, fontsize=10, ha="center", va="center", wrap=True)

            # Draw connector from level to steps
            ax.add_line(Line2D([0.5, x], [pos_y - 0.04, step_y + 0.04], color="black", linestyle="--"))

    # Finalize schematic
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis("off")
    plt.title("Map Construction Framework", fontweight='bold', fontsize=14, pad=20)
    plt.show()

create_clean_schematic()
```
